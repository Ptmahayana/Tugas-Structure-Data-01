```python
#Soal1 Input 10 Nilai
print("==============================================")
print("===== Program Menghitung Nilai Mahasiswa =====")
print("==============================================")
nilai_mhs = []
for i in range(10):
    nilai = int(input("Masukkan angka: "))
    nilai_mhs.append(nilai)
#Soal2 Nilai Tertinggi Dan Terendah
print("Nilai:", nilai_mhs)
print("Nilai maksimum:", max(nilai_mhs))
print("Nilai minimum:", min(nilai_mhs))
#Soal3 Rata - Rata
print("Rata-rata:", sum(nilai_mhs)/len(nilai_mhs))
#Soal4 Menghitung Berapa Total Mahasiswa Yang lulus Dan Tidak
lulus = 0
tidak_lulus = 0
def cek_lulus(nilai):
  if nilai >= 60:
    return True
  else:
    return False
for nilai in nilai_mhs:
  if cek_lulus(nilai):
    lulus += 1
  else:
    tidak_lulus += 1
print("Jumlah mahasiswa yang lulus:", lulus)
print("Jumlah mahasiswa yang tidak lulus:", tidak_lulus)
print("================================================")
#Soal5 Chart Nilai Terkecil Dan Terbesar
import matplotlib.pyplot as plt
nilai_tertinggi = max(nilai_mhs)
nilai_terendah = min(nilai_mhs)
plt.title('Nilai Mahasiswa')
nilai_ekstrim = [nilai_terendah, nilai_tertinggi]
labels = ['Terendah', 'Tertinggi']
plt.bar(labels, nilai_ekstrim, color=['yellow', 'green'])
plt.show()
print("==================================================")
#Soal6 Chart Jumlah Mahasiswa Yang Lulus Dan Tidak
mhs_lulus = lulus
mhs_tidak_lulus = tidak_lulus
sizes = jumlah_anggota = [mhs_lulus, mhs_tidak_lulus]
plt.title("Jumlah Mahasiswa Lulus Dan Tidak Lulus")
labels = ['Lulus', 'Tidak Lulus']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
jumlah_mhs = [mhs_lulus, mhs_tidak_lulus]
plt.axis('equal')
plt.show()
print("==================================================")
```
